# O que é prouni?
"O Programa Universidade para Todos (Prouni) é uma iniciativa do governo federal para facilitar o acesso de alunos de baixa renda ao ensino superior. Criado em 2004, o Prouni oferece bolsas de estudos parciais (50%) e integrais em faculdades privadas."

```python
# Importando bibliotecas importantes
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

Fazendo a leitura do csv escolhido e a leitura das primeiras informaçoes

```python
df = pd.read_csv('dados/prouni_2005_2019.csv')
df.head()
```

Explicando as variáveis:

- ANO_CONCESSAO_BOLSA: Ano que foi oferecida a bolsa.

- CODIGO_EMEC_IES_BOLSA: Codigo do EMEC

- NOME_IES_BOLSA: Universidade que foi oferecida a bolsa.

- TIPO_BOLSA: Verifica se a bolsa é integral ou parcial.

- MODALIDADE_ENSINO_BOLSA: Verifica se a modalidade é EAD ou PRESENCIAL.

- NOME_CURSO_BOLSA: Nome do curso ofertado.

- NOME_TURNO_CURSO_BOLSA: Turno do beneficiário da bolsa.

- CPF_BENEFICIARIO_BOLSA: Cpf do beneficiario

- SEXO_BENEFICIARIO_BOLSA: Sexo do beneficiário da bolsa.

- RACA_BENEFICIARIO_BOLSA: Etnia do beneficiário da bolsa.

- DT_NASCIMENTO_BENEFICIARIO: Data de nascimento do beneficiário da bolsa.

- BENEFICIARIO_DEFICIENTE_FISICO: Verifica se o beneficiário da bolsa é deficiente fisíco ou não.

- REGIAO_BENEFICIARIO_BOLSA: Região na qual reside o candidato que recebeu a bolsa.

- SIGLA_UF_BENEFICIARIO_BOLSA: UF no qual reside o candidato que recebeu a bolsa.

- MUNICIPIO_BENEFICIARIO_BOLSA: Municipio no qual reside o candidato que recebeu a bolsa.

- idade: A idade do candidato.

```python
# Irei retirar colunas que não influenciará na analise como: data de nascimento, cpf  
# removendo campo CPF do candidato
df_bolsas_2018.drop('CPF_BENEFICIARIO_BOLSA', axis=1, inplace=True)
# removendo data de nascimento  do candidato 
df_bolsas_2018.drop('DT_NASCIMENTO_BENEFICIARIO', axis=1, inplace=True)
```

```python
# Verificiando as colunas foram removidas
df_bolsas_2018.columns
# Vendo a nova quantidade de colunas
df_bolsas_2018.shape
print("A nova quantidade de colunas é: ", df_bolsas_2018.shape[1])
```

```python
# Fazendo a quantidade de beneficiarios por bolsa em cada estado no ano de 2018
count = df_bolsas_2018['SIGLA_UF_BENEFICIARIO_BOLSA'].value_counts()
print("\nBolsas concedidas em cada estado:\n",count)
```

```python
# Vendo apenas em SP
print("Bolsas concedidas em SP: ", count["SP"])
```

```python
#NOME_CURSO_BOLSA
count = df_bolsas_2018["NOME_CURSO_BOLSA"].value_counts()[:5]
count
```

Vendo cursos com maior oferta de bolsas 

```python
fig, ax = plt.subplots(figsize=(10, 8))
fig.suptitle("CURSOS COM MAIS CONCESSAO DE BOLSAS EM 2018")
sns.barplot(x=count.index, y=count, ax=ax, palette = "Set2")
plt.show()
```

Vemos que no ano de 2018 foram concedidas mais bolsas para o curso de direito

```python
count_tipo = df_bolsas_2018["TIPO_BOLSA"].value_counts()
count_tipo
```

Vemos que não há valores para bolsa complementar 25%, por isso removerei essa informaçao da tabela


```python
fig, ax = plt.subplots(figsize=(10, 8))
fig.suptitle("TIPOS DE BOLSAS MAIS CONCEDIDAS EM 2018")
sns.barplot(x=count_tipo.index, y=count_tipo, ax=ax, palette = "RdYlBu")
plt.show()
```

Vemos também que foram concedidas mais bolsas integrais do que parciais no ano de 2018

```python
# Aqui veremos quais etnias mais se beneficiaram com as bolsas
count_et = df_bolsas_2018.groupby('RACA_BENEFICIARIO_BOLSA') ['RACA_BENEFICIARIO_BOLSA'].count()
count_et
```

```python
fig, ax = plt.subplots(figsize=(10, 8))
fig.suptitle("Distribuição bolsas de acordo com a etnia")
plot = sns.barplot(x=count_et.index, y=count_et, ax=ax, palette = "Set2")
for i in plot.patches:
     plot.annotate(i.get_height(), 
    (i.get_x()+ i.get_width()/2, i.get_height()),
     ha="center",  va="baseline")
plt.show()
```

Notamos que aproximadamente 85% das bolsas foram concedidas para pessoas pardas e brancas

```python
# Analisando distribui;cao de bolsas por sexo
s_count = df_bolsas_2018.groupby("SEXO_BENEFICIARIO_BOLSA")["SEXO_BENEFICIARIO_BOLSA"].count()
print(s_count)
```

```python
colors = ['m','c']
fig1, ax1 = plt.subplots(figsize=(10,10))
plt.title("Distribuição bolsas de acordo com o sexo", color="white", size=15)
plt.pie(s_count, labels = s_count.index, autopct='%1.2f%%', textprops = dict(color ="w", size=12), shadow=True, startangle=180, colors=colors)
plt.show()
```

Vemos que as candidatas femininas possuem maior quantidade de bolsas 

```python
# contando as modalidades de ensino
count_mod = df_bolsas_2018['MODALIDADE_ENSINO_BOLSA'].value_counts()
# contando de ensino presencial
count_mod_pres = df_bolsas_2018[df_bolsas_2018['MODALIDADE_ENSINO_BOLSA'] == 'PRESENCIAL'].value_counts().sum()
# contando de ensino ead
count_mod_ead = df_bolsas_2018[df_bolsas_2018['MODALIDADE_ENSINO_BOLSA'] == 'EAD'].value_counts().sum()

print(f'Foram concedidas {count_mod_pres} na modalidade presencial e {count_mod_ead} na modalidade EAD')
```

```python
fig, ax = plt.subplots(figsize=(10, 8))
fig.suptitle("MODALIDADE DE ENSINO COM MAIS CONCESSAO DE BOLSAS EM 2018")
sns.barplot(x=count_mod.index, y=count_mod, ax=ax, palette = "RdYlBu")
plt.show()
```

Vemos que foram concecidadas mais bolsas do tipo presencial no ano de 2018

```python
# Farei a contagem de bolsas para cada universidade 
count_uni = df_bolsas_2018['NOME_IES_BOLSA'].value_counts()[:5]
count_uni
```

```python
# count uni - todos os anos
count_uni2 = df['NOME_IES_BOLSA'].value_counts()[:5]
count_uni2
```

```python
cor = ['salmon', 'green', 'blue', 'gold','plum']
plt.barh(count_uni.index, count_uni, color = cor)
plt.ylabel("UNIVERSIDADES")
plt.xlabel("DISTRIBUIÇÃO DAS BOLSAS")
plt.title("UNIVERSIDADES QUE MAIS CONCEDERAM BOLSAS NO ANO DE 2018")
plt.show()
```

```python
cor = ['salmon', 'green', 'blue', 'gold','plum']
plt.barh(count_uni2.index, count_uni2, color = cor)
plt.ylabel("UNIVERSIDADES")
plt.xlabel("DISTRIBUIÇÃO DAS BOLSAS")
plt.title("UNIVERSIDADES QUE MAIS CONCEDERAM BOLSAS NO PERIDO DE 2005-2019")
plt.show()
```

Percebemos que a Universidade Paulista(UNIP) é a instuição que mais ofertou bolsas, não só no ano de 2018 mas no período de 2005-2019.
